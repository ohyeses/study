```python
#지역 리스트 
location_list = ['서울','광주','부산','창원','인천','대전','대구','수원']
#테이블 연결 
cursor_weather = connection.cursor()
#table drop 
sql_droptable = "DROP TABLE WEATHER"
cursor_weather.execute(sql_droptable)
weatherloc = connection.commit()
#sequence drop
sql_dropsequence = "DROP SEQUENCE INDEX_NUM"
cursor_weather.execute(sql_dropsequence)
weatherloc = connection.commit()
#create sequence
sql_createsequence = "CREATE SEQUENCE INDEX_NUM START WITH 1 INCREMENT BY 1"
cursor_weather.execute(sql_createsequence)
weatherloc = connection.commit()
#create table 
sql_createtable = "CREATE TABLE WEATHER (INDEX_NUM NUMBER NOT NULL,LOCATION VARCHAR2(10), DAY VARCHAR2(10),TEMP NUMBER,HUMIDITY NUMBER, RAINPROB NUMBER,SKY NUMBER,CONSTRAINT weather_day_pk PRIMARY KEY(INDEX_NUM))"
cursor_weather.execute(sql_createtable)
weatherloc = connection.commit()

for i in range(0,8):
    location = location_list[i]
    #지역별로 7일치 날씨 넣기
    for j in range(0,7): 
        day = int(today) + j
        temp = data_list[i][0][j]
        humidity = data_list[i][1][j]
        rainprob = data_list[i][2][j]
        sky = data_list[i][3][j]
        
        if day > 20210331:
            day = day + 69
    
        sql_weather = "INSERT INTO WEATHER VALUES(INDEX_NUM.NEXTVAL,'{}','{}','{}','{}','{}','{}')".format(location,day,temp,humidity,rainprob,sky)
        cursor_weather.execute(sql_weather)
        weatherloc = connection.commit()
#DB 닫기
cursor_weather.close()
connection.close()       
```
