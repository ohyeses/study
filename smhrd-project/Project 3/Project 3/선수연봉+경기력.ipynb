```python
member = pd.read_csv("data/선수경기력1.csv", encoding = "utf-8")
member = member.set_index(["player_list"])
member
```

```python
baseball = pd.read_csv("data/선수연봉리스트.csv", encoding = "utf-8")
baseball = baseball.set_index(["player_list"])
baseball
```
