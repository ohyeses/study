```python
import pandas as pd
```

```python
data = pd.read_csv('Traffic_Accident_2017.csv', encoding = 'euc-kr')
data
```

```python
temp = data['요일'].value_counts()
temp
```

### 데이터의 순서를 월화수목금토일로 바꿔주기

```python
y = temp[['월', '화', '수', '목', '금','토', '일']]
y
```

```python
x = ['월','화', '수','목','금','토','일']
plt.bar(x,y)
plt.ylim(500, 620)
plt.xlabel('요일')
plt.ylabel('사고건수').set_rotation(0)
plt.title('요일별 사망교통사고 건수')
plt.show()
```

### 사고유형 대분류

#### 차대차 사건 중 사상자가 많은  지도를 시각화해보자

1. 사고유형 대분류 칼럼에서 차대차 사건만 불리언 인덱싱

```python
data.columns
```

#### 1642 rows x 27 columns

```python
car_to_car = data[data['사고유형_대분류']=='차대차']
car_to_car
```

```python
car_to_car[['사상자수','발생지시도']]
```

```python
car_to_car_by_place = car_to_car[['사상자수','발생지시도']].groupby('발생지시도').sum()
```

### 그래프 크기 키우기

```python
x = car_to_car_by_place.index
y = car_to_car_by_place.values
```

```python
plt.figure(figsize=(15,8)) # 그래프 크기 키우기
plt.plot(x,y)
plt.xlabel('지역')
plt.ylabel('사상자수')
plt.title('차vs차 교통사고의 사상자 수')
plt.show()
```
