### 전체 평점 평균 구하기

### 1번 사용자의 평점 평균 구하기
1. 1번 사용자 Boolean 인덱싱 해오기
2. 인덱싱 해온 데이터의 평점 평균 구하기

```python
#1단계
user_id1 = data[data[:,0] == 1]
```

```python
#2단계
user_id1[:,2].mean()
```

### 각 사용자 별 평점 평균 구하기
- for 문 이용
- unique() : 데이터에서 중복을 제거하고 나열해주는 함수

```python
user_id = np.unique(data[:,0]) # 1~6040번 까지 중복을 제거하고 하나씩만 출력해준다.
user_id
print(user_id.dtype)
```

```python
user_id.dtype
```

```python
score = []
```

```python
for i in user_id: # for문 한번 돌아갈 때 마다 i값을 가지고 있는 유저를 불러옴.
    user_id_score = data[data[:,0]==i][:,2].mean()
    score.append([i,user_id_score])
    
score_array = np.array(score).astype('float')
score_array
```

```python
# 쌤 풀이
for i in user_id:
    print(data[data[:,0]==i][:,2].mean()) # user id가 i인 유저들. ~6040번 까지
    
```

```python
data[data[:,0]==1][:,2].mean()  #1번 사용자의 평점의 평균
```
