### 새로운 값 추가하기 (검색과 비슷)
- 단가 컬럼 만들기
- 단가 = 가격/면적

```python
df['단가'] = df['가격']/df['면적']
df
```

```python
df['단가'] = df['단가'].astype('int') # 정수로 표현
df
```

- 가격, 면적, 단가 컬럼의 10번까지 출력

```python
df.loc[:10,('가격','면적','단가')]
# df.loc[:10,['가격', '면적', '단가']] 와 똑같음
```

### 데이터 정렬하기
- 가격을 오름차순으로 정렬하고 모든 데이터 출력

```python
df.sort_values(by = '가격')
```

- 가격을 오름차순으로 정렬한 후 가격과 지역 컬럼만 가져오기

```python
df.sort_values(by = '가격').loc[:10,['가격', '지역']]
```

- 4억이 넘는 아파트를 면적의 오름차순으로 정렬하고 가격,면적,지역 컬럼 가져오기

```python
df.sort_values(by = '면적')
```

```python
df[df['가격']>40000].sort_values(by='면적').loc[:,['가격','면적','지역']]
```
