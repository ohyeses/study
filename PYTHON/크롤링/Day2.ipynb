# 여러 페이지 리뷰 한번에 다 가져오기

```python
title_list = []
review_list = []

for i in range(1,5):

    url = "https://movie.naver.com/movie/point/af/list.nhn?&page=" + str(i)
    res = req.get(url)
    soup  = bs(res.text, 'lxml')

    title = soup.select('.color_b')
    review = soup.select('.title')

    a = soup.select('a')
    # div 만큼은 클래스명으로 가져와주쟈.다 지워짐
    div = soup.select('.list_netizen_score') 
    span = soup.select('st_off') 

    for i in a : 
        i.extract()

    for i in div : 
        i.extract()

    for i in span : 
        i.extract()


    review = soup.select('.title')


    for i in range(len(title)):
        title_list.append(title[i].text.strip())
        review_list.append(review[i].text.strip())
```

## tqdm_notebook >> 반복문의 진행 상황을 볼 수 있는 함수 

```python
from tqdm import tqdm_notebook
```

```python
title_list = []
point_list = []
rank_list = []
date_list = []
cnt = 1

for day in tqdm_notebook(days):
    url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=" + day

    res = req.get(url)
    soup = bs(res.text, 'lxml')

    result = soup.select('.tit5 > a')
    point = soup.select('.point')

    for i in range(len(result)):
        title_list.append(result[i].text)
        point_list.append(point[i].text)
        rank_list.append(i+1)
        date_list.append(day)
```

```python
dic = {'영화제목' : title_list, '평점' : point_list , '번호' : rank_list, '날짜' : date_list}
df2 = pd.DataFrame(dic)
df2.to_csv('./movie_rate.csv', encoding = 'euc-kr')
```
